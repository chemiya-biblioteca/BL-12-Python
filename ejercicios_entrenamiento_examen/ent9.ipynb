```python
import pandas as pd
```

```python
leido=pd.read_csv("boston.csv")
```

```python
leido
```

```python
leido=leido.dropna(subset=["MEDV"]).reset_index()
leido.shape
```

```python
X=leido.iloc[:,0:leido.shape[1]-1]
X
```

```python
y=leido["MEDV"]
y
```

```python
from sklearn.preprocessing import MinMaxScaler
minmax=MinMaxScaler()
X=minmax.fit_transform(X)
```

```python
from sklearn.preprocessing import LabelEncoder
valores=LabelEncoder().fit_transform(y)
valores
```

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.linear_model import SGDRegressor
skf=StratifiedKFold(n_splits=10)
```

```python
resultados=pd.DataFrame(columns=["fold","R2","aciertos"])
```

```python
#FSAR
k=1
for train_index,test_index in skf.split(X,valores):
    error=pd.DataFrame(columns=["original","prediccion","error","error relativo"])
    sgd=SGDRegressor(max_iter=1000)
    sgd.fit(X[train_index],y[train_index])
    predecido=sgd.predict(X[test_index])
    puntuacion=sgd.score(X[test_index],y[test_index])

   
    error["original"]=y[test_index]
    error["prediccion"]=predecido
    error["error"]=abs(error["original"]-error["prediccion"])
    error["error relativo"]=(error["error"]/error["original"])*100
    contador=0
    for h in range(len(error)):
        valor=error.iloc[h]["error relativo"]
        if(valor<15):
            contador+=1
    porcentaje=(contador/len(error))*100
    resultados.loc[k]=[k,puntuacion,porcentaje]
    k+=1
    
```
