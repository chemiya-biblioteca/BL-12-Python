```python
import pandas as pd
```

```python
datos=pd.read_csv("convertido.csv")
```

```python
datos
```

```python
y=datos["class"]
y
```

```python
X=datos.drop(columns=["class"])
X
```

```python
from sklearn.preprocessing import LabelBinarizer
import numpy as np
d=LabelBinarizer().fit(np.unique(y)).transform(y)
d
```

```python
from sklearn.preprocessing import MinMaxScaler
minmax=MinMaxScaler()
X=minmax.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,d_train,d_test=train_test_split(X,d,test_size=1/3)
```

```python
pred=np.zeros(d_test.shape)
pred
```

```python
from sklearn.linear_model import LinearRegression
```

```python
for i in range(d_test.shape[1]):
    
    lin=LinearRegression()
    lin.fit(X_train,d_train[:,i])
    pred[:,i]=lin.predict(X_test)
pred_fin=np.argmax(pred,axis=1)
pred_fin
    
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
print(accuracy_score(pred_fin,np.argmax(d_test,axis=1)))
print(confusion_matrix(pred_fin,np.argmax(d_test,axis=1)))
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import StratifiedKFold
```

```python
skf=StratifiedKFold(n_splits=10)
```

```python
res=pd.DataFrame(columns=["fold","aciertos"])
```

```python
k=0
for train_index,test_index in skf.split(X,y):
    log=LogisticRegression(solver="lbfgs",multi_class="auto")
    log.fit(X[train_index],y[train_index])
    prediccion=log.predict(X[test_index])
  
    res.loc[len(res)]=[k,accuracy_score(prediccion,y[test_index])]
    k+=1
```

```python
res
```

```python
print(res["aciertos"].mean())
```
