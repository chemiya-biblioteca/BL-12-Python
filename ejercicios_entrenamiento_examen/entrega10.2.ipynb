```python
from sklearn.datasets import fetch_openml
```

```python
X,y=fetch_openml(data_id=1560,return_X_y=True)
```

```python
X
```

```python
y
```

```python
import pandas as pd
import numpy as np
```

```python
from sklearn.preprocessing import MinMaxScaler
minmax=MinMaxScaler()
X=minmax.fit_transform(X)
```

```python
resultados=pd.DataFrame(index=range(2,8,2),columns=["entrena","test","tiempo"])
resultados.index.name="neuronas"
```

```python
resultados
```

```python
from sklearn.model_selection import StratifiedKFold
skf=StratifiedKFold(n_splits=10)
```

```python
from sklearn.neural_network import MLPClassifier
```

```python
import time
```

```python
for i in resultados.index:
    fold=pd.DataFrame(columns=["entrena","test","tiempo"])
    mlp=MLPClassifier(hidden_layer_sizes=i,max_iter=1000)
    k=0
    for train_index, test_index in skf.split(X,y):
        inicio=time.time()
        mlp.fit(X[train_index],y[train_index])
        fold.loc[k]=[mlp.score(X[train_index],y[train_index]),mlp.score(X[test_index],y[test_index]),time.time()-inicio]
        k+=1
    resultados.loc[i]=[fold.mean()[0],fold.mean()[1],fold.sum()[2]]
        
```
