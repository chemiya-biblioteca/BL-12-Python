```python
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LinearRegression
import pandas as pd
```

```python
leido=pd.read_csv("convertido.csv")
leido
```

```python
leido.shape
```

```python
X=leido.iloc[:,0:leido.shape[1]-1]
X
```

```python
y=leido.iloc[:,leido.shape[1]-1]
y
```

```python
from sklearn.preprocessing import LabelBinarizer
binarios=LabelBinarizer().fit_transform(y)
binarios
```

```python
from sklearn.preprocessing import LabelEncoder
valores=LabelEncoder().fit_transform(y)
valores
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,binarios,test_size=1/3)
```

```python
import numpy as np
predicciones_acumuladas=np.zeros(y_test.shape)
predicciones_acumuladas.shape
```

```python
from sklearn.metrics import accuracy_score
```

```python
for i in range(predicciones_acumuladas.shape[1]):
    lin=LinearRegression()
    lin.fit(X_train,y_train[:,i])
    prediccion=lin.predict(X_test)
    predicciones_acumuladas[:,i]=prediccion
prediccion_final=np.argmax(predicciones_acumuladas,axis=1)
reales=np.argmax(y_test,axis=1)
puntuacion=accuracy_score(prediccion_final,reales)
print(puntuacion)
```

```python
from sklearn.model_selection import StratifiedKFold
skf=StratifiedKFold(n_splits=10)
```

```python
#FLPA
resultados_log=pd.DataFrame(columns=["fold","aciertos"])
```

```python
from sklearn.preprocessing import MinMaxScaler
minmax=MinMaxScaler()
X=minmax.fit_transform(X)
```

```python
fold=1
for train_index,test_index in skf.split(X,valores):
    log=LogisticRegression(multi_class="auto",solver="lbfgs")
    log.fit(X[train_index],valores[train_index])
    puntuacion=log.score(X[test_index],valores[test_index])
    resultados_log.loc[len(resultados_log)]=[fold,puntuacion]
    fold+=1
    
    
    
    
```

```python
resultados_log
```

```python
resultados_log.aciertos.mean()
```

```python
#FYFLPA
resultados_fold=pd.DataFrame(columns=["fold","aciertos"])
k=1
for train_index,test_index in skf.split(X,y):
    y_valores=np.zeros(binarios[test_index].shape)
    for i in range(binarios.shape[1]):
        lin=LinearRegression()
        lin.fit(X[train_index],binarios[train_index,i])
        prediccion=lin.predict(X[test_index])
        y_valores[:,i]=prediccion
    maximos=np.argmax(y_valores,axis=1)
    puntuacion_acc=accuracy_score(maximos,valores[test_index])
    resultados_fold.loc[k]=[k,puntuacion_acc]
    k+=1
```
