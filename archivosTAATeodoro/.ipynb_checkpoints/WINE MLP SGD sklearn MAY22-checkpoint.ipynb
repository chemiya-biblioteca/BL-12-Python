```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_wine
from sklearn.preprocessing import Binarizer, MinMaxScaler
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
datos = load_wine()
```

```python
df = pd.DataFrame(columns=[datos.feature_names], data=datos.data)
df.insert(loc=len(df.columns), column="Class", value=datos.target)
df
```

```python
X = MinMaxScaler().fit_transform(np.array(df.iloc[:, :-1], dtype=float))
y = np.array(df.iloc[:,-1], dtype=int)
```

```python
EPOCHS = 50
#creo mlp
model = MLPClassifier(hidden_layer_sizes=(64,), solver='sgd', batch_size=1, 
                      max_iter=178*EPOCHS, learning_rate_init=0.01, verbose=True, tol=1e-8)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=1/3, stratify=y)
```

```python
model.fit(X_train, y_train)
```

```python
accuracy_score(model.predict(X_test), y_test)
```
