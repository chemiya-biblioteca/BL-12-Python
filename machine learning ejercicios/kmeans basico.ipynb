```python
import matplotlib.pyplot as plt

x = [4, 5, 10, 4, 3, 11, 14 , 6, 10, 12]
y = [21, 19, 24, 17, 16, 25, 24, 22, 21, 21]

plt.scatter(x, y)
plt.show()
```

```python
from sklearn.cluster import KMeans

data = list(zip(x, y))
#(4,21),(5,19)... junta x con y
inertias = []
```

```python
for i in range(1,11):
    kmeans = KMeans(n_clusters=i)#kmenas con los vecinos que toquen
    kmeans.fit(data)#ajusto el data en zip
    inertias.append(kmeans.inertia_)#anado la inertia que tenga

#inercia: suma de la distancia al cuadrado de cada muestra al centro del cluster mas cercano 
#poca inercia y pocos cluster es lo bueno
#num clusters donde la disminucion de la inertia es muy baja
```

```python
plt.plot(range(1,11), inertias, marker='o')
plt.title('Elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```

```python
kmeans = KMeans(n_clusters=2)
kmeans.fit(data)
#nos quedamos con 2 clusters
plt.scatter(x, y, c=kmeans.labels_)
plt.show()
```
