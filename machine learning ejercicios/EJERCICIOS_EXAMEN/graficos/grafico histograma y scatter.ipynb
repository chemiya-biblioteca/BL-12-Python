```python
import numpy
from scipy import stats#traigo de scipy
import matplotlib.pyplot as plt
```

```python
x = numpy.random.uniform(0.0, 5.0, 250)#250 muestas entre 0-5

plt.hist(x,5)# 5 grupos
plt.show()
```

```python
x = numpy.random.normal(5.0, 1.0, 100000)#meida 5, desv 1, 100000 muestras

plt.hist(x, 100)
plt.show()
```

```python
x = numpy.random.normal(5.0, 1.0, 1000)#media 5, desv 1, 1000 muestras
y = numpy.random.normal(10.0, 2.0, 1000)

plt.scatter(x, y)
plt.show()
```
